```python
weather = pd.read_csv('weather.csv')
weather.head()
```

```python
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],var_name='day',value_name='temp')
weather_melt
```

```python
weather_tidy=weather_melt.pivot_table(index=['id','year','month','day'],columns='element',values='temp').reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('output/weather.csv')
```

```python
weather_tidy.loc[(weather_tidy.month=='1') & (weather_tidy.day=='d15')]
```

```python
def avg(arr):
    return sum(arr)/len(arr)
```

```python
weather_tidy[['month','year']].apply(avg, axis=1).head()
```

```python
def extract(arr):
    return [int(dayStr[1:]) for dayStr in arr]
# extract day 12
```

```python
day = weather_tidy[['day']].apply(extract)
day.head()
```

```python
weather_tidy['day']=day
weather_tidy.head()
```
