```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = cv2.imread('/users/yajashkadian/data/landscape.jpg')
```

```python
(h,w,c) = im.shape
(h,w,c)
```

```python
image = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack([im,image]))
plt.show()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
image = image.reshape(h*w,3)
```

```python
numClusters = 4
clt = MiniBatchKMeans(n_clusters=numClusters)
labels = clt.fit_predict(image)
```

```python
print(labels.shape)
labels[:1000:100]
```

```python
clt.cluster_centers_.astype('uint8')
```

```python
quant = clt.cluster_centers_.astype('uint8')[labels]
```

```python
quant = quant.reshape((h,w,3))
```

```python
plt.imshow(quant)
plt.show()
```

```python
cv2.cvtColor(quant,cv2.COLOR_RGB2BGR)
cv2.imwrite('output,compressed.jpg',quant)
```
